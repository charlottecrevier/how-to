In this notebook, we are going to stream HRDEM Mosaic 1m data from Datacube for the bounding box of our area of interest. After that, we will use the pyflwdir library to calculate the flow direction on our HRDEM file.

### **About pyflwdir Library**


pyflwdir is a Python library designed for flow direction analysis. It allows us to compute the flow directions of water based on a Digital Elevation Model (DEM). This library is particularly useful for hydrological modeling and environmental studies. In our case, we will use pyflwdir to determine how water would flow across the terrain based on the elevation data provided by the HRDEM.

```python
import pyflwdir
import pystac_client
import stackstac
import numpy as np
import rasterio
import matplotlib.pyplot as plt
```

```python
catalog        = pystac_client.Client.open('https://datacube.services.geo.ca/api')  
stac_items     = catalog.search(collections = ['hrdem-mosaic-1m'], 
                                bbox = [-116.225, 58.3172, -115.83, 58.5178], 
                                datetime = '2019-01-01/2024-12-31'
                               )
stac_found_items = [item for item in stac_items.items()]
```

```python
def reorder_transform(gdal_transform):
    return [gdal_transform[1], gdal_transform[2], gdal_transform[0],  # x scale (a), rotation about y (b), x-coordinate of upper left (c)
            gdal_transform[4], gdal_transform[5], gdal_transform[3],  # rotation about x (d), y scale (e), y-coordinate of upper left (f)
            0, 0, 1]  # The last part is fixed for affine transformation matrix (0, 0, 1)


for item in stac_found_items:
    if "proj:transform" in item.properties:
        if isinstance(proj_transform, list):
            item.properties["proj:transform"] = reorder_transform(item.properties["proj:transform"])
    else:
        print(f"Item {item.id} does not have proj:transform in properties.")
```

```python
xrDS = stackstac.stack(stac_found_items[0], 
                          assets        = ["dtm"], 
                          resolution    = 1.0,
                          bounds_latlon = bbox, 
                          chunksize     = (1000, 1000))
```

So far, we have streamed our HRDEM using the stack.satc library, as we discussed in the previous notebook. At this point, we now have a Dask array, not a NumPy array. However, the pyflwdir library works with NumPy arrays! This raises the question: **how can we leverage the advantages of both Dask and pyflwdir at the same time?**

Dask provides a function called **map_overlap**. This function allows us to define another function, such as the flow_direction function, that we want to apply to all the chunks of our data in parallel. **Chunks of data?** Remember that we are working with a Dask array, and Dask divides the data into smaller chunks that can be processed independently.

In map_overlap, we can run a function on all the chunks in parallel, which is where the power of Dask comes into play. The cool thing about map_overlap is that it works with NumPy arrays. When we call the flow_direction function in map_overlap, Dask will pass each chunk of data as a NumPy array to the function. This means we can define a function that works with NumPy arrays, like the flow_direction function from the pyflwdir library.

The key takeaway is that the output of these functions must also be NumPy arrays. Dask will then handle the parallel processing of the chunks, and once all the chunks are processed, it will combine them—like solving a puzzle—to give us the final full array. This allows us to take advantage of both Dask’s parallel processing and pyflwdir's ability to work with NumPy arrays.

- 




Now, let's code the function that we want to run on our chunks of data (or, technically, on our data arrays). In the following function, we receive a NumPy array block_data, which has dimensions (time, y, x), and also a crs argument, which is the coordinate reference system (CRS) of our dataset.

In this function, we call the flow direction calculation on each chunk of data, making sure the chunk is not filled with no-data values. We also ensure that null cells are properly handled. If the chunk contains no valid data (i.e., it is entirely filled with no-data values), we return a chunk filled with null values, but in a two-dimensional array structure.

```python
def flow_dir(block_data,latlon):
    
    if not np.all(np.isnan(block_data[0, :, :])):
        
        # Create flow direction from DEM data
        flw = pyflwdir.from_dem(
            data = block_data[0, :, :],
            nodata = np.nan,
            latlon=latlon
        )
        # Convert the flow data to an array
        flow_array = flw.to_array()
        flow_array = flow_array.astype(float)
        flow_array[np.isnan(block_data[0, :, :])] = np.nan

        return flow_array
    else:
        
        return np.full((block_data.shape[1], block_data.shape[2]), np.nan)
    
```

Now, we need to call the flow_dir function on each chunk of data in our Dask array. To do this, we use map_overlap on our Dask array and define the function. The extra arguments (not the chunk itself) include the depth, which is the overlap of chunks in the x and y dimensions, and the chunksize for the returned data array.

We are calling map_**overlap** because the flow direction calculation relies on the neighboring cells for each cell. To ensure that the edges of the chunks, which may fall at any parts of the big dask array, have access to neighboring cells, we need to provide overlaps between adjacent chunks. The cool thing is that these overlaps will be handled automatically when the chunks are processed.

Thus, we benefit from access to neighboring cells without modifying the overall structure of the array.

```python
flow_direction = dtm_dask.map_overlap(flow_dir, 
                                      latlon=xrDS.rio.crs, 
                                      depth = {"x": 1000, "y": 1000}, 
                                      chunks = (1000, 1000), drop_axis=0, dtype =np.float32)
```
